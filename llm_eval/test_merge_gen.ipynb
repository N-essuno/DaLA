```python
import pandas as pd
```

```python
# local_generations = "logs/comma-v0.1-1t_log.csv"

# local_generations = "logs/llama-3.8b-instruct_log.csv"
# euroeval_generations = "logs/dala-gen-dataset_predictions_custom_llama_inst_all.csv"

euroeval_generations = "logs/dala-gen-dataset_predictions_custom_llama_base.csv"
local_generations = "logs/llama-3-8b_log.csv"

output_path = "logs/updated_csv.csv"
```

```python
local_gen_df = pd.read_csv(local_generations)
euroeval_gen_df = pd.read_csv(euroeval_generations)

# drop duplicates based on "target_text" column
euroeval_gen_df = euroeval_gen_df.drop_duplicates(subset=["target_text"])

# remove ' and " from start and end of text and generation columns in euroeval_gen_df
euroeval_gen_df["text"] = euroeval_gen_df["text"].str.strip("'\"")
euroeval_gen_df["generation"] = euroeval_gen_df["generation"].str.strip("'\"")

print(f"N rows CSV 1: {len(local_gen_df)}")
print(f"N rows CSV 2: {len(euroeval_gen_df)}")

if len(local_gen_df) != len(euroeval_gen_df):
    raise Exception("Row number mismatch")

# # merge adding columns from euroeval_gen_df to local_gen_df
# merged_df = pd.concat([local_gen_df, euroeval_gen_df], axis=1)

# merge dfs on columns "target_text_2" from local_gen_df and "target_text" from euroeval_gen_df
merged_df = pd.merge(local_gen_df, euroeval_gen_df, left_on="target_text_2", right_on="target_text", how="inner")

# save merged df to csv with ; as separator
merged_df.to_csv(output_path, index=False, sep=';')
```
